## **Hypothesis Testing**

```python
mu = 1
sigma = 1
data_normal = np.random.normal(mu,sigma,100)
data_array = np.array(data_normal)
normalized_data = data_array / np.linalg.norm(data_array)
print(normalized_data)
```

```python
from matplotlib import pyplot as plt
plt.plot(normalized_data,color = 'green')
plt.show()
```
