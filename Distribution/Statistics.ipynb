## **Relative Frequency**

```python
cars_in_country = pd.crosstab(df["origin"],"Number of cars")
```

```python
cars_in_country["relatice_frequency"] = cars_in_country["Number of cars"]/cars_in_country["Number of cars"].sum()
```

```python
cars_in_country
```

```python
cars_in_country["percentage"] = cars_in_country["Number of cars"]/cars_in_country["Number of cars"].sum()*100
```

```python
cars_in_country
```
