```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = sns.load_dataset("mpg")
```

```python
data
```

```python
sns.barplot(x = 'cylinders', data = data)
```

```python
plt.plot(data['mpg'],data['cylinders'])
plt.plot(data['mpg'],data['displacement'])
plt.legend()
plt.xlabel('mpg')
plt.ylabel('cylinders')
plt.title('mpg vs cylinders')
plt.show()
```
