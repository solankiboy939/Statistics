```python
import numpy as np
import pandas as pd
import scipy
```

```python
alpha = 0.5
```

```python
def ztest(n,sample_mean,pop_mean,pop_sd,alternative):
  z_score = (sample_mean-pop_mean)/(pop_sd/np.sqrt(n))
  p_value = scipy.stats.norm.sf(z_score)
  print(z_score)
  print(p_value)
  if(alternative == "one-sided"):
    if(p_value < alpha):
      print("reject null hypothesis")
    else:
      print("accept null hypothesis")
  elif(alternative == "two-sided"):
    if(p_value < alpha/2):
      print("reject null hypothesis")
    else:
      print("accept null hypothesis")
```

```python
ztest(50,110,100,15,alternative="one-sided")
```

```python
ztest(81,90,82,20,"one-sided")
```

# ***One-Way ANOVA***

```python
from scipy.stats import f_oneway
```

```python
e = [2,4,2]
m = [2,3,4]
s  = [1,2,5]
```

```python
stat,pvalue = f_oneway(e,m,s)
```

```python
alpha = 0.05
```

```python
if(pvalue < alpha):
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```

```python
performace1 = [80,89,88,78,79]
performace2 = [93,92,94,89,88]
performace3 = [89,76,34,56,91]
performace4 = [40,64,89,28,39]
```

```python
stat ,pvalue = f_oneway(performace1,performace2,performace3,performace4)
```

```python
if(pvalue < alpha):
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
```
