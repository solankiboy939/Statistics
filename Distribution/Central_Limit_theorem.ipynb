```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
sns.displot(df['displacement'], kind = 'kde')
```

```python
sample_mean_list = []
for i in (0,30):
  random_sample = df['displacement'].sample(n = 30)
  # sample_mean = random_sample.mean()
  sample_mean_list.append(random_sample.mean())
```

```python
sns.displot(sample_mean_list, kind = 'kde')
```

```python
import statistics
statistics.mean(sample_mean_list)
# sample_mean_list
```
